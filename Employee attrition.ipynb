### Data Preprocessing

#### Convert Attrition from ('Yes', 'No') to (1,0) 

```python
def Attritrion(n):
    if n=='Yes':
        return 1
    else:
        return 0
```

```python
df['Attrition']=df['Attrition'].apply(Attritrion)
```

```python
df.head()
```

### Convert the rest of the categorical values into numeric using dummy variables and store the results in a new dataframe called 'newdf'

## Machine Learning Models

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.model_selection import cross_val_score
```

```python
reg=LogisticRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
prediction=reg.predict(X_test)
```

```python
print('Accuracy score of our model is :',metrics.accuracy_score(y_test,prediction))
print('Confusion matrix is :')
print(metrics.confusion_matrix(prediction,y_test))
print('Classification report is:')
print(metrics.classification_report(y_test,prediction))
print('Cross Validation Score is:',cross_val_score(reg,X,y,cv=6).mean())
```

<span style = "color:orange"> Visualize confusion matrix </span>

```python
sns.heatmap(metrics.confusion_matrix(prediction,y_test),annot=True,fmt='d')
```

### Random Forest Classifier
** Choose the best estimator and parameters :GridSearchCV**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
forest=RandomForestClassifier()
```
